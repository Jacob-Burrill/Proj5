```python
#Importing necessary packages
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

The visualization that I will be attempting to recreate can be found at this link: https://ourworldindata.org/military-personnel-spending. The data set describes military spending by country since 1949. Not all countries have data available from every year.

```python
df = pd.read_csv("military-expenditure-share-gdp.csv")
df.info()
```

```python
df.head(20)
```

Like the visualization I am recreating, I am limiting the countries that I will display to France, Russia, China, United Kingdom, and United States. This will help declutter the graph.  

```python
df_France = df[df['Entity']== 'France']
df_US = df[df['Entity']== 'United States']
df_UK = df[df['Entity']== 'United Kingdom']
df_China = df[df['Entity']== 'China']
df_Russia = df[df['Entity']== 'Russia']
```

```python
plt.plot(df_France['Year'], df_France['military_expenditure_share_gdp'], label = "France", color = 'orange')
plt.plot(df_US['Year'], df_US['military_expenditure_share_gdp'], label = "United States", color = 'salmon')
plt.plot(df_Russia['Year'], df_Russia['military_expenditure_share_gdp'], label = "Russia", color = 'pink')
plt.plot(df_UK['Year'], df_UK['military_expenditure_share_gdp'], label = "United Kingdom", color = 'forestgreen')
plt.plot(df_China['Year'], df_China['military_expenditure_share_gdp'], label = "China", color = 'navy')
plt.xlabel("Year")
plt.ylabel("Military Expenditure")
plt.title("Military expenditure as a share of GDP, 1949 to 2020")
plt.legend()
plt.show()
```
